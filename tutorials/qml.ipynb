Quantum machine learning (QML) is a rapidly evolving field that combines quantum computing with machine learning techniques. One of the key challenges in QML is how to effectively encode classical data into quantum states so that quantum algorithms can process it. Then, we need to craft quantum circuits to work as a quantum neural network. In this notebook, we will explore some of the encoding methods used in QML, including basis encoding, amplitude encoding, and angular encoding.

We can have an overview of how does a quantum neural network works in the following figure:

We notice the following steps in the figure:
1. **Data Loading**: Classical data is loaded into the quantum computer. This is typically done through encoding techniques that map classical data into quantum states as we have seen in the previous section. Also, it can be encoded with smart ways like using different feature maps or encoding methods implemented in Qiskit, such as `ZFeatureMap` or `PauliFeatureMap`.
2. **Quantum Ansatz**: A quantum circuit is designed to represent the quantum neural network. This circuit is parameterized, meaning that it can be adjusted based on the data and the task at hand. The ansatz is a crucial part of the quantum neural network, as it determines how the quantum state will evolve based on the input data.
There are different types of ansatzes implemented in Qiskit, such as `EfficientSU2`, `RealAmplitudes`, and `TwoLocal`. Each of these ansatzes has its own structure and can be used for different types of problems. The choice of ansatz can significantly affect the performance of the quantum neural network.
3. **Measurement**: After the quantum circuit has processed the input data, the quantum state is measured. This measurement collapses the quantum state into a classical bitstring, which can then be used for further processing or classification.

The training part: Once the quantum neural network is set up, it can be trained using classical optimization techniques. The parameters of the quantum circuit are adjusted based on the training data to minimize a loss function, which measures how well the quantum neural network performs on the task at hand. This process is similar to training classical neural networks, but it leverages the unique properties of quantum computing to potentially achieve better performance. #### Data Encoding in Quantum Machine Learning: Basis encoding 

In quantum machine learning, data encoding is a crucial step that allows classical data to be represented in quantum states. One of the simplest methods for encoding classical data into quantum states is **basis encoding**. 

Imagine having a classical dataset with features that can take on discrete values. For example, the distance versus time dataset of a car's journey traveling between two cities. 

Example dataset:

| Distance (km) | Time (hours) |
|---------------|--------------|
| 2           | 1            |
| 5           | 2            |
| 7           | 3            |
| 8           | 4            |
| 10           | 5           |

This dataset contains two features: distance and time. Each feature can take on values from a finite set, such as integers from 1 to 5 in this case.§

Each feature in this dataset can be represented as a quantum state, where the values of the features correspond to the basis states of qubits.

In basis encoding, each feature is represented by a quantum bit (qubit) in its computational basis state. For example, if you have a feature that can take on values 0 or 1, you can represent it as a qubit in the |0⟩ or |1⟩ state.
However, if the feature can take on more than two values, you can use multiple qubits to represent it. 

To encode, we need to map the classical data to quantum states. Basis encoding means that we represent each value in the spin or the bit information of the qubit. 

First, we need to convert them to binary. Based on the needed digits of the binary representation, we can determine how many qubits we need. For example, if we have values between distance 0 and 10, we need at least 4 qubits to represent the values in binary (0000 to 1010).

Hence, for the distance feature, we can represent the values as follows:
- 0 km: |0000⟩
- 1 km: |0001⟩
- 2 km: |0010⟩
- 3 km: |0011⟩
- 4 km: |0100⟩
- 5 km: |0101⟩
- 6 km: |0110⟩
- 7 km: |0111⟩
- 8 km: |1000⟩
- 9 km: |1001⟩
- 10 km: |1010⟩

### Amplitude Encoding
Amplitude encoding is another method for encoding classical data into quantum states. In amplitude encoding, the classical data is represented in the amplitudes of a quantum state. This allows for a more compact representation of the data, as it can encode multiple values in a single quantum state. We will not cover this method in detail here, but you can find more information in the [IBM tutorial](https://quantum.cloud.ibm.com/learning/en/courses/quantum-machine-learning/data-encoding).

### Angular Encoding
Angular encoding is another method for encoding classical data into quantum states. In angular encoding, the classical data is represented in the angles of a quantum state. This allows for a more compact representation of the data, as it can encode multiple values in a single quantum state. We will not cover this method in detail here, but you can find more information in the [IBM tutorial](https://quantum.cloud.ibm.com/learning/en/courses/quantum-machine-learning/data-encoding).
